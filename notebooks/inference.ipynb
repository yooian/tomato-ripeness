```python
import tensorflow as tf

# Load the exported model directory
inference_model = tf.keras.models.load_model("../models/updated_tomato_ripeness_model.keras")
```

```python
import numpy as np
from tensorflow.keras.applications.efficientnet_v2 import preprocess_input

# Load and preprocess the image
img_path = "../images/PICT0012.JPG"
img = tf.keras.utils.load_img(img_path, target_size=(256, 256))
img_array = tf.keras.utils.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
img_array = preprocess_input(img_array)  # Must match training preprocessing
```

```python
predictions = inference_model.predict(img_array)
predicted_class = np.argmax(predictions, axis=1)[0]
print(f"Predicted class index: {predicted_class}")

class_names = ['Damaged', 'Old', 'Ripe', 'Unripe']
print(f"Predicted label: {class_names[predicted_class]}")

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Read the image using matplotlib.image.imread()
image = mpimg.imread(img_path)

# Display the image using matplotlib.pyplot.imshow()
plt.imshow(image)

# Remove axis ticks and labels for cleaner visualization (optional)
plt.xticks([])
plt.yticks([])

# Show the plot
plt.show()
```
